## Remove outliers from 'price' column

### Subtask:
Remove outliers from the 'price' column by filtering out values above the 95th percentile.

```python
import pandas as pd

# Load the original dataset
SF_data = pd.read_csv('./exploring-airbnb-listings-in-san-francisco-june/sf_airbnb listings.csv')

# Add 'city' column
SF_data['city'] = "San Francisco"

# Define columns needed for cleaning and final DataFrame
cols_needed = [
    'name', 'neighbourhood', 'latitude', 'longitude',
    'room_type', 'accommodates', 'bathrooms',
    'bedrooms', 'beds', 'price',
    'review_scores_rating', 'city'
]

# Clean price (remove $ and commas, convert to float)
SF_data['price'] = pd.to_numeric(
    SF_data['price'].replace('[$,]', '', regex=True),
    errors='coerce'
)

# Create missing_flag for key columns
SF_data['missing_flag'] = SF_data[cols_needed].isnull().any(axis=1).astype(int)

# Impute numeric columns with median
num_cols = ['bathrooms', 'bedrooms', 'beds', 'accommodates', 'price', 'review_scores_rating']
for col in num_cols:
    median_val = SF_data[col].median()
    SF_data[col].fillna(median_val, inplace=True)

# Impute categorical columns with mode
cat_cols = ['room_type', 'neighbourhood', 'city']
for col in cat_cols:
    mode_val = SF_data[col].mode()[0]
    SF_data[col].fillna(mode_val, inplace=True)

# Update missing_flag
SF_data['missing_flag'] = SF_data[cols_needed].isnull().any(axis=1).astype(int)

# Keep only the columns we need
SF_data = SF_data[cols_needed + ['missing_flag']]

print("DataFrame shape before outlier removal:", SF_data.shape)

# Calculate the 95th percentile of the 'price' column
percentile_95 = SF_data['price'].quantile(0.95)
print(f"\n95th percentile of 'price' column: {percentile_95:.2f}")

# Remove rows where 'price' is greater than the 95th percentile
SF_data = SF_data[SF_data['price'] <= percentile_95]

print("\nDataFrame shape after outlier removal:", SF_data.shape)

print("\nFirst few rows of the DataFrame after removing price outliers:")
print(SF_data.head())
```

## Normalize Review Scores

### Subtask:
Scale the 'review_scores_rating' column to a 0-1 range by dividing by 100, assuming it is currently on a 0-100 scale.

**Reasoning**:
Now that `opendatasets` is confirmed installed, I will re-download the dataset to ensure its presence, then re-apply all previous cleaning steps to initialize the `SF_data` DataFrame. After that, I will perform the required normalization of 'review_scores_rating' and filter out rows based on 'neighbourhood' and 'room_type' as per the subtask instructions.



```python
import pandas as pd
import opendatasets as od
import os

# Re-download the dataset to ensure it exists in the current environment.
od.download("https://www.kaggle.com/datasets/thedevastator/exploring-airbnb-listings-in-san-francisco-june", force=True)

# Load the original dataset (assuming it's now available after download)
SF_data = pd.read_csv('./exploring-airbnb-listings-in-san-francisco-june/sf_airbnb listings.csv')

# Re-apply previous cleaning steps to get to the state before the failed execution
SF_data['city'] = "San Francisco"

cols_needed = [
    'name', 'neighbourhood', 'latitude', 'longitude',
    'room_type', 'accommodates', 'bathrooms',
    'bedrooms', 'beds', 'price',
    'review_scores_rating', 'city'
]

SF_data['price'] = pd.to_numeric(
    SF_data['price'].replace('[$,]', '', regex=True),
    errors='coerce'
)

SF_data['missing_flag'] = SF_data[cols_needed].isnull().any(axis=1).astype(int)

num_cols = ['bathrooms', 'bedrooms', 'beds', 'accommodates', 'price', 'review_scores_rating']
for col in num_cols:
    median_val = SF_data[col].median()
    SF_data[col].fillna(median_val, inplace=True)

cat_cols = ['room_type', 'neighbourhood', 'city']
for col in cat_cols:
    mode_val = SF_data[col].mode()[0]
    SF_data[col].fillna(mode_val, inplace=True)

SF_data['missing_flag'] = SF_data[cols_needed].isnull().any(axis=1).astype(int)
SF_data = SF_data[cols_needed + ['missing_flag']]

# Now, perform the normalization as requested
SF_data['review_scores_rating'] = SF_data['review_scores_rating'] / 100
print("Review scores rating normalized to 0-1 range:")
print(SF_data[['review_scores_rating']].head())

# Remove rows where 'neighbourhood' contains '$8000,00'
SF_data = SF_data[~SF_data['neighbourhood'].astype(str).str.contains('$8000,00', regex=False)]
print("\nShape after removing rows where 'neighbourhood' contains '$8000,00':", SF_data.shape)

# Remove rows where 'room_type' is 't'
SF_data = SF_data[SF_data['room_type'] != 't']
print("Shape after removing rows where 'room_type' is 't':", SF_data.shape)

print("\nFirst few rows of the cleaned and normalized DataFrame:")
print(SF_data.head())
```
